# Quantitative Results

In this section we will process the mAP50 results obtained from the experiments

## Generic Functions

```python
import json
import matplotlib.pyplot as plt
import numpy as np

def read_json_as_dict(file_path):
        """
        Reads a JSON file and returns its contents as a dictionary.
        
        Parameters:
        file_path (str): The path to the JSON file.

        Returns:
        dict: The contents of the JSON file as a dictionary.
        """
        with open(file_path, 'r') as file:
            data = json.load(file)
        return data

def plot_map(mAP_values, cls=None):
    epochs = list(range(1, len(mAP_values)+1))
    plt.plot(epochs, mAP_values, linestyle='-', color='b')
    plt.xlabel('Iteration number')
    plt.ylabel('mAP50 Value')
    if cls is not None:
        plt.title(f'mAP50 Evolution for {cls} category')
    else:
        plt.title('mAP50 Evolution')
    plt.grid(True)
    plt.show()
    
def print_hyperparameters(results_dict, index):
    text = ''
    for module in results_dict['Hyperparameters']:
        for param in results_dict['Hyperparameters'][module]:
            if isinstance(results_dict['Hyperparameters'][module][param][index], list):
                gammalut_str = ', '.join(f'{value:.2f}' for value in results_dict['Hyperparameters'][module][param][index])
                text += '- {}/{}: [{}] \n'.format(module, param, gammalut_str)
            else:
                text += '- {}/{}: {:.2f} \n'.format(module, param, results_dict['Hyperparameters'][module][param][index])

    print(f'The best hyperparameters combination with mAP of {results_dict["mAP"][index]*100:.2f}% are: \n{text}')

def show_results(results, space_reduction_iterations):
    map50 = results["mAP"]
    plot_map(map50[space_reduction_iterations:])
    max_mAP50 = np.max(map50)*100
    print(f'The maximum mAP achieved is {max_mAP50:.2f} %')
    print(f'The final mAP achieved is {map50[-1]*100:.2f} %')
    
    max_map_index = np.argmax(map50)

    print_hyperparameters(results, max_map_index)
    
    
```

- Hyperparameters optimized: SET 1	
- Classes: Car, motorbike, person

```python
results_01 = read_json_as_dict("results_01.json")
space_reduction_iterations = 100

show_results(results_01, space_reduction_iterations)
```
